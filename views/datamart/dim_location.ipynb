# dim_location

Locatie-dimensie met SHA2 surrogate key. NULL `location_id` (warn-rule in Silver) collapst op één gedeelde "Unknown"-key.

Georkestreerd door `views/07_apply_views.ipynb`.

```python
%sql
CREATE WIDGET TEXT catalog DEFAULT "DEMO";
USE CATALOG ${catalog};

CREATE OR REPLACE VIEW DATAMART.dim_location AS
SELECT
  SHA2(COALESCE(CAST(location_id AS STRING), '__UNKNOWN__'), 256) AS dim_location_key,
  location_id
FROM INTEGRATION.order_header
GROUP BY location_id;
```
