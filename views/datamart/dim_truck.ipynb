# dim_truck

Truck-dimensie met SHA2 surrogate key. NULL `truck_id` (warn-rule in Silver) collapst op één gedeelde "Unknown"-key.

Georkestreerd door `views/07_apply_views.ipynb`.

```python
%sql
CREATE WIDGET TEXT catalog DEFAULT "DEMO";
USE CATALOG ${catalog};

CREATE OR REPLACE VIEW DATAMART.dim_truck AS
SELECT
  SHA2(COALESCE(CAST(truck_id AS STRING), '__UNKNOWN__'), 256) AS dim_truck_key,
  truck_id
FROM INTEGRATION.order_header
GROUP BY truck_id
ORDER BY truck_id;
```
