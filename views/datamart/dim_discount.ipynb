# dim_discount

Korting-dimensie met SHA2 surrogate key. `discount_id` is nullable; orders zonder korting collapsen op één gedeelde "Unknown"-key.

Georkestreerd door `views/07_apply_views.ipynb`.

```python
%sql
CREATE WIDGET TEXT catalog DEFAULT "DEMO";
USE CATALOG ${catalog};

CREATE OR REPLACE VIEW DATAMART.dim_discount AS
SELECT
  SHA2(COALESCE(CAST(discount_id AS STRING), '__UNKNOWN__'), 256) AS dim_discount_key,
  discount_id
FROM INTEGRATION.order_header
GROUP BY discount_id;
```
